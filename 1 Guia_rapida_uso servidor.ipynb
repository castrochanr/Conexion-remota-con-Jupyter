## mover archivos


### _*mv*_

Primero hay que saber que la accin de  _**mover**_ y _**transferir**_ no son lo mismo, transferir consiste de la gestion de archivos y directorios de un equipo local a uno remoto y viceversa.

El comando _**mv**_ es equivalente a copiar y pegar en una sola accin y en esta nica accin hay que definir las rutas completas de origen y destino de los archivos a copiar.

Si escribimos el comando man y la orden cp en la terminal, nos dir como utilizar. man est disponible para muchos de los programas de bash.

```python
man mv
```

### _cp_ 

#### Comando _cp_ aplicado a archivos

*para archivos la sintaxis debe ser como se muestra a continuacin*:

<kbd>__cp__</kbd>    <kbd>__ORIGEN__</kbd>    <kbd>__DESTINO__</kbd>


```python
cp /home/rcastro/001_000.jpg /home/rcastro/Imgenes/
```

Si yo ejecuto el comando escrito en la lnea anterior, le estara dando la orden a la terminal de copiar un archivo de la ruta /home/user/ a la ruta /home/user/Imgenes/, al poner el nombre y extension del archivo \*.jpg, slo en el origen, entonces no se cambiar el nombre del archivo a copiar _001_000.jpg_.

### _cp_

#### Este es el mismo comando mencionado arriba, pero aplicado a directorios

*para diectorios se requiere la opcin _"-R"_ que en realidad signfica que el proceso sea reiterativo y la sintaxis debe ser como se muestra a continuacin*:

<kbd>__cp__</kbd>    <kbd>__-R__</kbd> <kbd>__ORIGEN__</kbd>    <kbd>__DESTINO__</kbd>


```python
cp -R PATH_ORIGEN PATH_DESTINO # es reiterativo para que copie todo el contenido interno
```

## Transferir archivos


Esto bsicamente consiste de enviar, archivos de un recurso local a uno remoto.


### _scp_

Este comando, *scp*, requiere del origen y el destino en los equipos local y remoto, estas rutas servirn para gestionar los archivos mediante ssh y scp.

para indicar el manual debemos escribir _man scp_

```python
man scp #nos muestra el manual de scp
```

Dado que un comando tiene cierta sintaxis:
    
    comando <opciones> <parmetro1> <parmetro2> ...

*scp* se gestiona desde el recurso en el cual se haya la terminal interactuando.

Si la terminal se encuentra conectada al servidor, entonces scp se puede usar

```python
scp [options] ORIGEN DESTINO
```

donde:

origen es el archivo que quiere copiar y destino es la ruta a donde lo va a copiar, por ejemplo:

#### _Enviar_

La operacin a continuacin la podramos definir como enviar.

Si tenemos un servidor con direccin ip 192.168.8.1, entonces para copiar un archivo de un equipo local a uno remoto escribimos:

```python
scp archivo_local user@192.168.8.1:/home/user/PATH_destino/archivo
```

#### _Importar_

Pero si lo que deseamos hacer es copiar un archivo del servidor remoto a la maquina local entonces lo que debemos ejecutar ser algo similar a:


```python
scp user@192.168.8.1:/home/user/PATH_destino/archivo PATH_local/archivo_local
```

El ejemplo anterior podramos definirlo como importar.


__Nota: Si se incluye la opcin -R  antes del origen, el proceso se convierte en reiterativo y puede ejecutarse sobre directorios__
